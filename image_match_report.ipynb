```python
files = glob.glob('./data/editions/*.xml')
print(len(files))
```

```python
graphic_url = []
for x in files:
    doc = TeiReader(x)
    for graphic in doc.any_xpath('.//tei:graphic'):
        graphic_url.append(graphic.attrib['url'])
```

```python
ref_img_names = [x.replace('anno:', '') for x in graphic_url]
print(len(ref_img_names))
```
